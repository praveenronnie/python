```python
def oddOrEven(num):
    if(num % 2 == 0):
        return "even"
    else:
        return "Odd"

value = int(input("Enter the number: "))
print(oddOrEven(value))
```

```python
def findPercentage(marks):
    return sum([x for x in marks])/ len(marks)

marks = []
totalSubjects = int(input("Enter total number of Subjects: "))
for i in range (1, totalSubjects + 1):
    sub = int(input(f"Subject {i} = "))
    marks.append(sub)

print(sum(marks))
print(findPercentage(marks))

"""
Output:
Subject1= 98
Subject2= 87
Subject3= 95
Subject4= 95
Subject5= 93
Total : 468
Percentage : 93.60000000000001
"""
```

```python
"""
Height:32
Breadth:34
Area formula: (Height*Breadth)/2
Area of Triangle: 544.0
Height1:2
Height2:4
Breadth:4
Perimeter formula: Height1+Height2+Breadth
Perimeter of Triangle: 10
"""

def areaOfTraingle(height, breadth):
    return (height * breadth) / 2

def perimeterOfTraingle(h1, h2, b):
    return h1 + h2 + b

height = float(input("Height: "))
breadth = float(input("Breadth: "))
area = areaOfTraingle(height, breadth)
print("Area of Traingle: " +  str(area))

h1 = float(input("Height1: "))
h2 = float(input("Height2: "))
b = float(input("Breadth: "))
perimeter = perimeterOfTraingle(h1, h2, b)
print("Perimeter of Traingle: " +  str(perimeter))
```
